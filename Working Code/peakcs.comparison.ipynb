```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv("~/Downloads/Solar_flare_RHESSI_2004_05.csv")

data1['date'] = pd.to_datetime(data1[['year', 'month']].assign(day=1))

mean_peak_count_rate = data1.groupby(['year', 'month'])['peak.c/s'].mean().unstack()


data_2004 = mean_peak_count_rate.loc[2004]
print("Mean Peak Count Rate for Each Month in 2004:")
print(data_2004)

data_2005 = mean_peak_count_rate.loc[2005]
print("Mean Peak Count Rate for Each Month in 2005:")
print(data_2005)


# plot for 2004
plt.figure(figsize=(8, 6))
data_2004.plot(kind='bar', color='skyblue', edgecolor='skyblue')
plt.title('Mean Peak Count Rate in 2004')
plt.xlabel('Month')
plt.ylabel('Mean Peak Count Rate')
plt.show()


# plot for 2005
plt.figure(figsize=(8, 6))
data_2005.plot(kind='bar', color='salmon', edgecolor='salmon')
plt.title('Mean Peak Count Rate in 2005')
plt.xlabel('Month')
plt.ylabel('Mean Peak Count Rate')
plt.show()

```
