Computing Basic Stats for Data Set 1 and Data Set 2 comparison

```python
import pandas as pd

data1 = pd.read_csv("Solar_flare_RHESSI_2004_05.csv")
data2 = pd.read_csv("Solar_flare_RHESSI_2015_16.csv")

stats_df1 = data1.describe()
stats_df2 = data2.describe()


print("Dataset 1 statistics:")
print(stats_df1.to_string())
print()

print("Dataset 2 statistics:")
print(stats_df2.to_string())
```

Histogram for Frequency

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Solar_flare_RHESSI_2004_05.csv')

df['date'] = pd.to_datetime(df[['year', 'month']].assign(DAY=1))

months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['month'] = pd.Categorical(df['date'].dt.month_name(), categories=months_order, ordered=True)

# plt.figure(figsize=(12, 6))
# colors = ['skyblue', 'salmon']
# years_order = sorted(df['date'].dt.year.unique())

# for i, year in enumerate(years_order):
#     data_year = df[df['date'].dt.year == year]
#     data_year['month'].value_counts().sort_index().plot(kind='bar', label=str(year), alpha=0.7, color=colors[i])

# plt.title('Histogram of Months')
# plt.xlabel('Month')
# plt.ylabel('Frequency')
# plt.legend(title='Year')
# plt.show()
data = df['date']
mean_val = data.mean()
std_deviation = data.std()
print(f'Mean: {mean_val}')
print(f'Standard Deviation: {std_deviation}')

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

colors = ['skyblue', 'salmon']
years_order = sorted(df['date'].dt.year.unique())

for i, year in enumerate(years_order):
    data_year = df[df['date'].dt.year == year]
    data_year['month'].value_counts().sort_index().plot(kind='bar', ax=axs[i], label=str(year), alpha=0.7, color=colors[i])
    
axs[0].set_title('Occurence of Flares (2004)')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Frequency')

axs[1].set_title('Occurence of Flares (2005)')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Frequency')

plt.legend(title='Year', loc='upper left', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Solar_flare_RHESSI_2004_05.csv')


energy_bins = [6, 12, 25, 50, 100, 300, 800, 7000, 20000]
energy_labels = ['6-12', '12-25', '25-50', '50-100', '100-300', '300-800', '800-7000', '7000-20000']

df['energy_range'] = pd.cut(df['energy.kev.i'], bins=energy_bins, labels=energy_labels, include_lowest=True)

plt.figure(figsize=(10, 6))
df['energy_range'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Energy Ranges')
plt.xlabel('Energy Range (keV)')
plt.ylabel('Frequency')
plt.show()
```
